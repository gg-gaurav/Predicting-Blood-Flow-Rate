```python
# Split dataset into features and targets
X = dnew_fx.drop(['SpecificSigma'], axis = 1)
for column in X.columns:
    if X[column].dtypes == 'complex128':
        X[column] = X[column].abs()
y = dnew_fx['SpecificSigma'].abs()
```

```python
# Apply information gain
ig = mutual_info_regression(X,y)
```

```python
# Create dictionary of feature importance scores
feature_scores = {}
i = 0
for column in X.columns:
    feature_scores[column] = ig[i]
    i+=1
```

```python
# Sort feature score by importance
sorted_features = sorted(feature_scores.items(), key=lambda x: x[1], reverse=True)
sorted_features = sorted_features[:20]

# Print the feature importance scores and the sorted features
for feature, score in sorted_features:
    print("Feature:", feature, "Score:", score)
```

```python
# Plot a horizontal bar chart of the feature importance scores
fig, ax = plt.subplots(figsize = (15,10))
y_pos = np.arange(len(sorted_features))
ax.barh(y_pos, [score for feature, score in sorted_features], align="center")
ax.set_yticks(y_pos)
ax.set_yticklabels([feature for feature, score in sorted_features])
ax.invert_yaxis()  # Labels read top-to-bottom
ax.set_xlabel("Importance Score")
ax.set_title("Feature Importance Scores (Information Gain)")

# Add importance scores as labels on the horizontal bar chart
for i, v in enumerate([score for feature, score in sorted_features]):
    ax.text(v + 0.01, i, str(round(v, 3)), color="black", fontweight="bold")
plt.show()
```

## Flow as of now: Use Spearmann Correlation to find dependency between features... drop the highly correlated ones... then use KBest or Mutual Information find dependency with target val and drop low value ones

## Step 1: Spearman's Correlation

```python
featureSet_spearman = featureSet
Y_spearman = Y 
for column in featureSet_spearman.columns:
    if featureSet_spearman[column].dtypes == 'complex128':
        featureSet_spearman[column] = featureSet_spearman[column].abs()
```

```python
spearman = featureSet_spearman.corr(method = 'spearman')
spearman
```

```python
spearman_select = np.full((spearman.shape[0]) ,True ,dtype = bool)

threshold = 0.999
for i in range(spearman.shape[0]):
    for j in range(i+1, spearman.shape[0]):
        if (spearman.values[i,j] >= threshold) or (spearman.values[i,j] <= -threshold):
            if spearman_select[j]:
                spearman_select[j] = False


spearmanFeatures = spearman.columns[spearman_select]
featureSet_spearman = featureSet_spearman[spearmanFeatures]

featureSet_spearman
```

```python
spearman[spearmanFeatures]
```
